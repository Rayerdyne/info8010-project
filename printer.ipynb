```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def make_fig(s):
    avgs = pd.read_csv(s)
    plt.figure()
    plt.plot(avgs["Epoch number"], avgs["Average test loss"])
    plt.plot(avgs["Epoch number"], avgs["Average train loss"])
    plt.xlabel("Epoch")
    plt.ylabel("MSE Loss")
    plt.legend(["Average test loss", "Average train loss"])
    plt.show()
```

```python
make_fig("results/averages-2.csv") 
```

```python
def make_fig_grads(s):
    data = pd.read_csv(s)

    fig, ax = plt.subplots()
    xx = data["Parameter"]
    yy = data["Gradient"]
    plt.plot(xx, yy)
    plt.xlabel("Parameter index")
    plt.ylabel("Gradient L_2 norm")
    ax.set_ybound(0, 30)
    plt.legend(["Gradient norm"])
    plt.show()
```

```python
make_fig_grads("results/gradients-2.csv")
make_fig_grads("results/gradients-2-normalized.csv")
```
